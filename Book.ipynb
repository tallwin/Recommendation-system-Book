```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import the datasets
df=pd.read_csv("/content/book.csv",encoding='ISO-8859-1')
df
```

```python
book=df.drop ('Unnamed: 0',axis=1)
book
```

```python
#sort by user id
book.sort_values(['User.ID'])
```

```python
#rename the columns
book.rename(columns={'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'},inplace=True)
book
```

```python
book.isnull().sum()
```

```python
#number of unique user in the data set
len(book.userid.unique())
```

```python
#number of unique user in the data set
len(book.booktitle.unique())
```

```python
#convert into pivot table
new_book=book.pivot_table(index='userid',
    columns='booktitle',
    values='bookrating').reset_index(drop=True)
new_book
```

```python
#replacing the index values by unique user ids
new_book.index=book['userid'].unique()
new_book
```

```python
#Replace NaNs with 0s
new_book.fillna(0,inplace=True)
new_book
```

```python
#import libraries
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#Calculating Cosine Similarity between Users
user_sim = 1 - pairwise_distances( new_book.values,metric='cosine')
user_sim
```

```python
#store the results in dataframe
user_sim1=pd.DataFrame(user_sim)
user_sim1
```

```python
#Set the index and column names to user ids 
user_sim1.index = book.userid.unique()
user_sim1.columns = book.userid.unique()
user_sim1
```

```python
#Null diagonal values
np.fill_diagonal(user_sim, 0)
user_sim1
```

```python
#most similar users
user_sim1.idxmax(axis=1)
```
